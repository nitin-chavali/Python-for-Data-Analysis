```python
#create a line plot
x=[1,2,3,4,5]
y=[1,4,6,9,25]
plt.plot(x,y)
```

```python
#create a customized line plot.

plt.plot(x,y,color='red',linestyle=':',marker='*')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title("Basic Line plot")
plt.grid(True)
plt.show()
```

```python
#multiple plot
#Sample data
x=[1,2,3,4,5]
y1=[1,4,5,8,9]
y2=[10,20,30,40]

plt.figure(figsize=(5,1))
plt.subplot(1,2,1)
plt.plot(x,y1,color='green')
plt.title("PLOT 1")

plt.subplot(2,2,4)
plt.plot(x,y1,color='yellow')
plt.title("PLOT 1")
```

```python
#Bar plot

categories=['A','B','C','D']
Values=[5,6,7,8]
plt.figure(figsize=(5,2))
plt.bar(categories,Values,color='purple')
```

```python
#Histogram

data=[1,2,2,3,3,3,4,5,5]
plt.hist(data,bins=5,color='orange',edgecolor='black')
```

```python
#scatter plot

x=[1,2,4,6]
y=[3,4,6,8]
plt.scatter(x,y,color='blue',marker='o',edgecolor='red')
```

```python
#Pie chart
labels=['A','B','C','D']
sizes=[30,25,45,10]
colors=('grey','orange','green','yellow')
explode=(0.2,0,0,0)
plt.pie(sizes,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True,explode=explode)
```
